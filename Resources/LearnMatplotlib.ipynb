```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Type of plots

#### Line plots

```python
x = [1,2,3,4,5]
y = [10, 15, 12, 18, 20]
```

```python
plt.plot(x, y);
plt.title('Simple Line Plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True)
```

### Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'Jan']
revenue = [100, 200, 300, 500, 600]
plt.figure(figsize=(10, 6))
plt.bar(months, revenue);
plt.title('Monthly Revenue Bar Chart');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
```

#### Scatterplots

## Subplots and legends

```python
days = [1,2,3,4,5]
original_price = [10, 20, 25,30, 40]
predicted_price = [8, 21, 25, 28, 36]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

## Subplots

```python
# First subplot - Original price
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original Prices overtime')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Second subplot - Predicted price
plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Original Prices overtime')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Third subplot - Difference between Original and Predicted
price_difference = [o-p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, price_difference, color='purple', label='Price Difference')
plt.title('Differnce between Original and Predicted Prices')
plt.xlabel('Days')
plt.ylabel('Price Difference')
plt.legend()
plt.tight_layout()
```

#### Saving the chart

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.savefig('example_prediction.png', dpi=300)
```
